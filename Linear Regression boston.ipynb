# Linear Regression

## Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading boston data 

```python
from sklearn.datasets import load_boston #dataset loaded
```

```python
df = load_boston()
df
```

## Creating dataframe

```python
pd.DataFrame(df.data)
```

```python
dataset = pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names #adding the feature names as columns
```

```python
dataset.head()
```

```python
X = dataset #dependent variable
X
```

```python
Y = df.target #independent jo op me chahiye
Y
```

## Performing Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

## standardizing the dataset

```python
#standardazing the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(X_train)
```

```python
x_test = scaler.transform(X_test)
#to inverse
#scaler.iverse_transform(X_test)
```

## Importing Linear Regression Model

```python
from sklearn.linear_model import LinearRegression 
#cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(x_train,Y_train)
```

```python
mse = cross_val_score(regression , x_train , Y_train , scoring='neg_mean_squared_error' , cv=5)
```

```python
np.mean(mse)
```

```python
#prediction
reg_pred = regression.predict(X_test)
reg_pred
```

## Displot

```python
import seaborn as sns
sns.displot(reg_pred-Y_test,kind ='kde')
```

## R2Score

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_pred,Y_test)
```
